```python
# Student Performance Predictor

## Objective: To predict student marks based on study hours, sleep hours, and attendance using machine learning.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#create dataset

data = {
    'StudyHours': [2, 3, 5, 7, 8, 6, 4, 9, 1, 5],
    'SleepHours': [5, 6, 7, 6, 8, 7, 5, 8, 4, 6],
    'Attendance': [60, 65, 80, 85, 90, 75, 70, 95, 50, 78],
    'Marks': [50, 55, 70, 80, 90, 75, 65, 95, 45, 72]
}

df = pd.DataFrame(data)
df
```

```python
# data processing
X = df[['StudyHours', 'SleepHours', 'Attendance']]
y = df['Marks']
```

```python
#train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#model training
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#predictions on test data
predictions = model.predict(X_test)
print("Predictions:", predictions)

from sklearn.metrics import mean_squared_error

error = mean_squared_error(y_test, predictions)
print("Mean Squared Error:", error)
```

```python
#custom input prediction
sample = pd.DataFrame({
    'StudyHours': [6],
    'SleepHours': [7],
    'Attendance': [85]
})

result = model.predict(sample)

print("Predicted Marks:", result[0])
```

```python
#conclusion and markdown
plt.scatter(df['StudyHours'], df['Marks'])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```
